```python
import pandas as pd

path = "master_clean_slang.csv"

data = pd.read_csv(path)
#data.drop(data["american_indian"])
print(data.head(5))
```

```python
tweets_count = list(data['tweet_text'])
```

```python
tweet_dict= {}
for tweet in tweets_count:
    tweet = str(tweet)
    words = tweet.split()
    for word in words:
#         print(word)
        if(word not in tweet_dict.keys()):
            tweet_dict[word] = 0
        else:
            tweet_dict[word] += 1

for word in tweet_dict.keys():
    if(tweet_dict[word] >= 100):
        print(word, ":", tweet_dict[word])
```

```python
for word in tweet_dict.keys():
    if(word.startswith("HASH") and tweet_dict[word] >= 1000):
        print(word, ":", tweet_dict[word])
```

```python
vulgarities = ["shit","damn","ass","bitch","fuck","fucking","fucked","bullshit","hell"]
words_of_interest = ["stfu","smh","asap","chillin","wat","xd","af","ugh","sis","wtf"]
hashtags = ["HASHgetalljobs","HASHjobs","HASHwinning"]
```

```python
master_dict = {}
for word in tweet_dict.keys():
    if word in vulgarities:
        master_dict.update({word:tweet_dict[word]})
    elif word in words_of_interest:
        master_dict.update({word:tweet_dict[word]})
    elif word in hashtags:
        master_dict.update({word:tweet_dict[word]})
```

```python
master_df = [master_dict]
master_df = pd.DataFrame(master_df, index=['Frequency']).T
master_df
```
