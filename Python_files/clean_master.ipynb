```python
import pandas as pd
```

```python
path = "/Users/sarikasabnis/Documents/GitHub/Graphics_final/data/master_clean_slang.csv"

data = pd.read_csv(path)
data.head(5)
```

```python
vulgarities = ["shit","damn","ass","bitch","fuck","fucking","fucked","bullshit","hell"]
words_of_interest = ["stfu","smh","asap","chillin","wat","xd","af","ugh","sis","wtf"]
hashtags = ["HASHgetalljobs","HASHjobs","HASHsorrynotsorry","HASHwinning"]
```

```python
tweets = list(data["tweet_text"])
tweets[:10]
```

```python
tweet_dict = {}
for tweet in tweets:
    tweet = str(tweet)
    words = tweet.split()
    for word in words:
        if(word not in tweet_dict.keys()):
            tweet_dict[word] = 0
        else:
            tweet_dict[word] += 1
```

```python
tweet_dict
```

```python
master_dict = {}
for word in tweet_dict.keys():
    if word in vulgarities:
        master_dict.update({word:tweet_dict[word]})
    elif word in words_of_interest:
        master_dict.update({word:tweet_dict[word]})
    elif word in hashtags:
        master_dict.update({word:tweet_dict[word]})
```

```python
df = [master_dict]
df = pd.DataFrame(df, index=["2011","2012"]).T
```
